```python
import pandas as pd
```

```python
flow=pd.read_csv('./flow_betw_country.csv',index_col=0)
```

```python
country=flow['ques_owner_country'].to_list()
```

```python
diff={}
```

```python
for i in country:
    diff[i]={}
    for j in country:
        expor=flow[(flow['ans_owner_country']==i)&(flow['ques_owner_country']==j)]['count'].iloc[0]
        impor=flow[(flow['ans_owner_country']==j)&(flow['ques_owner_country']==i)]['count'].iloc[0]
        diff[i][j]=round((int(expor)-int(impor))/int(expor),3)
```
